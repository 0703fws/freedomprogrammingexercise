```python
#Read/Summarise Data
# import python packages
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# read the data from the provided url
url ="https://raw.githubusercontent.com/0703fws/freedomprogrammingexercise/refs/heads/main/paygap.csv"
data = pd.read_csv(url)
# show the dataset
display(data)
def summarise_data(data):
    # check dataset details
    data.info()
    # summary statistics for numerical columns
    display(data.describe())
    # count missing values
    print(data.isnull().sum())
# check if data is loaded successfully
if data is not None:
    summarise_data(data)
else:
    print("data loading failed")
```

```python
#ã€€Data Manipulation
# check if missing values position matches between male and female quartiles
overlap1 = data["male_lower_quartile"].isnull().equals(data["female_lower_quartile"].isnull())
overlap2 = data["male_lower_middle_quartile"].isnull().equals(data["female_lower_middle_quartile"].isnull())
overlap3 = data["male_upper_middle_quartile"].isnull().equals(data["female_upper_middle_quartile"].isnull())
overlap4 = data["male_top_quartile"].isnull().equals(data["female_top_quartile"].isnull())
# show matching results for each pair
overlap_result = {
    "male_lower_quartile vs female_lower_quartile":overlap1,
    "male_lower_middle_quartile vs female_lower_middle_quartile":overlap2,
    "male_upper_middle_quartile vs female_upper_middle_quartile":overlap3,
    "male_top_quartile vs female_top_quartile":overlap4
}
for group,match_result in overlap_result.items():
    print(f"{group}:{match_result}")
# define columns for each quartile
quartile_cols = ["male_lower_quartile","female_lower_quartile",
          "male_lower_middle_quartile","female_lower_middle_quartile",
          "male_upper_middle_quartile","female_upper_middle_quartile",
          "male_top_quartile","female_top_quartile"]
# fill missing values with median
for column in quartile_cols:
  data[column] = data[column].fillna(data[column].median())
# verify the sum of male-female pairs equals 100
pairs = [("male_lower_quartile","female_lower_quartile"),
    ("male_lower_middle_quartile","female_lower_middle_quartile"),
    ("male_upper_middle_quartile","female_upper_middle_quartile"),
    ("male_top_quartile","female_top_quartile")
]
for male,female in pairs:
  sum_pairs = data[male]+data[female]
  print(f"{male}+{female}:")
  print(f"sum:{sum_pairs.mean():.2f}")
# define columns for bonus
numeric_cols = ["diff_mean_bonus_percent","diff_median_bonus_percent"]
# fill with median to minimize impact on distribution
for column in numeric_cols:
  data[column] = data[column].fillna(data[column].median())
# drop irrelevant columns with missing values
data_cleaned = data.drop(columns=["company_link_to_gpg_info","responsible_person","post_code","address","company_number","sic_codes"])
# count latest missing values
print(data_cleaned.isnull().sum())
# print latest dataset size
print(f"data_cleaned shape:{data_cleaned.shape}")
```

```python
# Data visualisation: Pay gap by company size
# select columns for analysis
paygap_column = ["diff_mean_hourly_percent","diff_median_hourly_percent"]
# calculate mean pay gap grouped by company size
diff_size = data_cleaned.groupby("employer_size")[paygap_column].mean()
# create a bar chart
bar_chart = diff_size.plot(kind="bar",figsize=(10,6),edgecolor="black",width=0.8)
# add xlabel, ylabel and title
plt.xlabel("Company size",fontsize=12)
plt.ylabel("Pay gap percentage(%)",fontsize=12)
plt.title("Pay gap by company size",fontsize=18)
# rotate x-axis labels 45 degrees for easier reading and align them to the right
plt.xticks(rotation=45,ha="right")
# add legend and position outside the plot
plt.legend(["Mean Hourly Pay Gap","Median Hourly Pay Gap"],loc="upper left",bbox_to_anchor=(1,1))
# add percentage labels for each bar
for bar_group in bar_chart.containers:
  bar_chart.bar_label(bar_group,fmt="%.1f%%",label_type="edge",fontsize=8)
# adjust layout
plt.tight_layout()
# show the bar chart
plt.show()
```

```python
# Data visualisation: Gender distribution by pay quartiles
# define quartile level labels
quartile_labels = ["Lower Quartile","Lower Middle Quartile","Upper Middle Quartile","Top Quartile"]
# create male quartile dictionary
male_quartile = {"male_lower_quartile":data_cleaned["male_lower_quartile"],
          "male_lower_middle_quartile":data_cleaned["male_lower_middle_quartile"],
          "male_upper_middle_quartile":data_cleaned["male_upper_middle_quartile"],
          "male_top_quartile":data_cleaned["male_top_quartile"]
}
# calculate mean values for male in each quartile
male_quartile_mean = [male_quartile["male_lower_quartile"].mean(),
     male_quartile["male_lower_middle_quartile"].mean(),
     male_quartile["male_upper_middle_quartile"].mean(),
     male_quartile["male_top_quartile"].mean()
]
# create female quartile dictionary
female_quartile = {"female_lower_quartile":data_cleaned["female_lower_quartile"],
          "female_lower_middle_quartile":data_cleaned["female_lower_middle_quartile"],
          "female_upper_middle_quartile":data_cleaned["female_upper_middle_quartile"],
          "female_top_quartile":data_cleaned["female_top_quartile"]
}
# calculate mean values for female in each quartile
female_quartile_mean = [female_quartile["female_lower_quartile"].mean(),
     female_quartile["female_lower_middle_quartile"].mean(),
     female_quartile["female_upper_middle_quartile"].mean(),
     female_quartile["female_top_quartile"].mean()
]
# set the size of the figure
plt.figure(figsize=(10,6))
# create a line chart for male quartile mean
plt.plot(quartile_labels,male_quartile_mean,label="Male",marker="o",linestyle="-",linewidth=2) # set circle markers, solid line, width 2
# add percentage labels for male data points
for i,value in enumerate(male_quartile_mean):
  plt.annotate(f"{value:.1f}%",(quartile_labels[i],value),textcoords="offset points",xytext=(0,6),ha="center") # offset above point
# create a line chart for female quartile mean
plt.plot(quartile_labels,female_quartile_mean,label="Female",marker="o",linestyle="-",linewidth=2) # set circle markers, solid line, width 2
# add percentage labels for female data points
for i,value in enumerate(female_quartile_mean):
  plt.annotate(f"{value:.1f}%",(quartile_labels[i],value),textcoords="offset points",xytext=(0,-12),ha="center") # offset below point
# add xlabel, ylabel and title
plt.xlabel("Quartiles",fontsize=12)
plt.ylabel("Percentage(%)",fontsize=12)
plt.title("Gender Distribution Across Quartiles",fontsize=18)
# add legend
plt.legend()
# show the line chart
plt.show()
```

```python
# Data visualisation: Proportion of employees receiving bonus by gender
# calculate mean bonus percentage
bonus_mean = data_cleaned[["male_bonus_percent","female_bonus_percent"]].mean()
# set the size of the figure
plt.figure(figsize=(10,6))
# create a pie chart with percentage labels for each segment
plt.pie(bonus_mean,labels=["Male","Female"],autopct="%1.1f%%") # display percentage labels with one decimal place
# add title
plt.title("Proportion of Employees Receiving Bonus by Gender",fontsize=16)
# show the pie chart
plt.show()
```

```python
# Data visualisation: Gender pay gap distribution
# set the size of the figure
plt.figure(figsize=(10,6))
# create a histogram
plt.hist([data_cleaned["diff_mean_hourly_percent"],data_cleaned["diff_median_hourly_percent"]],bins=50,label=["Mean","Median"])
# add xlabel, ylabel and title
plt.xlabel("Pay Gap Percentage (%)",fontsize=12)
plt.ylabel("Number of Companies",fontsize=12)
plt.title("Distribution of Mean and Median Gender Pay Gap",fontsize=18)
# add legend
plt.legend()
# add grid
plt.grid(True,linestyle="--",alpha=0.5)  # set dashed lines and 50% transparency

# count extreme cases for mean pay gaps
mean_high = data_cleaned[data_cleaned["diff_mean_hourly_percent"] > 150]["diff_mean_hourly_percent"].count()
mean_low = data_cleaned[data_cleaned["diff_mean_hourly_percent"] < -100]["diff_mean_hourly_percent"].count()
# count extreme cases for median pay gaps
median_high = data_cleaned[data_cleaned["diff_median_hourly_percent"] > 150]["diff_median_hourly_percent"].count()
median_low = data_cleaned[data_cleaned["diff_median_hourly_percent"] < -100]["diff_median_hourly_percent"].count()
# get total number of companies
companies = len(data_cleaned)
# show the percentages of companies with extreme pay gaps (mean and median)
print(f"Percentage of companies with extreme gaps:")
print(f"Mean hourly gap > 150%:{(mean_high/companies*100):.4f}%")
print(f"Mean hourly gap < -100%:{(mean_low/companies*100):.4f}%")
print(f"Median hourly gap > 150%:{(median_high/companies*100):.4f}%")
print(f"Median hourly gap < -100%:{(median_low/companies*100):.4f}%")
# check extreme values in mean and median hourly pay gaps:
# The two columns contain extreme values, but their quantities are negligible and do not affect the visualization results.
# The x-axis range can be confirmed and adjusted to make the graph better.

# set x-axis ticks from -100 to 150 with an interval of 10
plt.xticks(np.arange(-100, 151, 10))
# set the x-axis display range from -100 to 150
plt.xlim(-100, 150)
# show the histogram
plt.show()
```

```python
# Data visualisation: Relationship between pay gap and bonus gap
# create a scatter plot
sns.scatterplot(data_cleaned,x="diff_mean_hourly_percent",y="diff_mean_bonus_percent")
# add a regression line without scatter points
sns.regplot(data_cleaned,x="diff_mean_hourly_percent",y="diff_mean_bonus_percent",scatter=False,color="orange",line_kws={"linestyle":"--"})
# add xlabel, ylabel and title
plt.xlabel("Mean Hourly Pay Gap (%)",fontsize=12)
plt.ylabel("Mean Bonus Gap (%)",fontsize=12)
plt.title("Relationship between Pay Gap and Bonus Gap",fontsize=18)
# add grid
plt.grid(True,linestyle="--",alpha=0.5) # set dashed lines and 50% transparency
# shoew the scatter plot
plt.show()
```

```python
# Data visualisation: Gender proportion gap across pay quartiles
# calculate gender proportion gap for each quartile
proportion_gap = [(m - f) for m,f in zip(male_quartile_mean,female_quartile_mean)]
# set the heatmap data
heatmap_data = pd.DataFrame({"Quartile":quartile_labels,"Proportion Gap":proportion_gap}).set_index("Quartile")
# set the size of the figure
plt.figure(figsize=(10,5))
# create a heatmap
sns.heatmap(heatmap_data.T, # transpose data for correct orientation
      annot=True, # show the data values inside the heatmap cells
      fmt=".1f", # display percentage labels with one decimal place
      cmap="RdYlBu_r", # use reversed "Red-Yellow-Blue" color map
      center=0, # center the color scale at zero for better visualisation
      cbar_kws={"label":"Proportion Gap (%)"}) # add label to the color bar
# add xlabel and title
plt.xlabel("Pay Quartiles",fontsize=12)
plt.title("Gender Proportion Gap Across Pay Quartiles",fontsize=18)
#show the heatmap
plt.show()
```

```python
# Data visualisation: Annual trend of gender pay gap
# convert date to datetime
data_cleaned["date_submitted"] = pd.to_datetime(data_cleaned["date_submitted"])
# create year column
data_cleaned["year"] = data_cleaned["date_submitted"].dt.year
# calculate annual average pay gap
yearly_gap = data_cleaned.groupby("year")["diff_mean_hourly_percent"].agg(["mean","count"]).reset_index()
# set the size of the figure
plt.figure(figsize=(10,6))
# create a line chart
plt.plot(yearly_gap["year"],yearly_gap["mean"],marker="*",markersize=8,linewidth=2)
# add value labels for each point
for x,y in zip(yearly_gap["year"],yearly_gap["mean"]):
  plt.text(x,y-0.1,f"{y:.1f}%",ha="center")
# add xlabel, ylabel and title
plt.xlabel("Year",fontsize=12)
plt.ylabel("Mean Hourly Pay Gap (%)",fontsize=12)
plt.title("Annual Trend of Gender Pay Gap",fontsize=18)
# add grid
plt.grid(True, linestyle="--",alpha=0.5)
# adjust layout
plt.tight_layout()
# show the line chart
plt.show()
```
