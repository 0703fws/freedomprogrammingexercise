```python
#Read/summarise data
# import python packages
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# read the data from the provided url
url ="https://raw.githubusercontent.com/0703fws/freedomprogrammingexercise/refs/heads/main/paygap.csv"
data = pd.read_csv(url)
# show the dataset
display(data)
def summarise_data(data):
    # check dataset details
    data.info()
    # summary statistics for numerical columns
    display(data.describe())
    # count missing values
    print(data.isnull().sum())
# check if data is loaded successfully
if data is not None:
    summarise_data(data)
else:
    print("data loading failed")
```

```python
#ã€€Data manipulation
# check if missing values position matches between male and female quartiles
overlap1 = data["male_lower_quartile"].isnull().equals(data["female_lower_quartile"].isnull())
overlap2 = data["male_lower_middle_quartile"].isnull().equals(data["female_lower_middle_quartile"].isnull())
overlap3 = data["male_upper_middle_quartile"].isnull().equals(data["female_upper_middle_quartile"].isnull())
overlap4 = data["male_top_quartile"].isnull().equals(data["female_top_quartile"].isnull())
# show matching results for each pair
overlap_result = {
    "male_lower_quartile vs female_lower_quartile":overlap1,
    "male_lower_middle_quartile vs female_lower_middle_quartile":overlap2,
    "male_upper_middle_quartile vs female_upper_middle_quartile":overlap3,
    "male_top_quartile vs female_top_quartile":overlap4
}
for group,match_result in overlap_result.items():
    print(f"{group}:{match_result}")
# define columns for each quartile
quartile_cols = ["male_lower_quartile","female_lower_quartile",
          "male_lower_middle_quartile","female_lower_middle_quartile",
          "male_upper_middle_quartile","female_upper_middle_quartile",
          "male_top_quartile","female_top_quartile"]
# fill missing values with median
for column in quartile_cols:
  data[column] = data[column].fillna(data[column].median())
# verify the sum of male-female pairs equals 100
pairs = [("male_lower_quartile","female_lower_quartile"),
    ("male_lower_middle_quartile","female_lower_middle_quartile"),
    ("male_upper_middle_quartile","female_upper_middle_quartile"),
    ("male_top_quartile","female_top_quartile")
]
for male,female in pairs:
  sum_pairs = data[male]+data[female]
  print(f"{male}+{female}:")
  print(f"sum:{sum_pairs.mean():.2f}")
# define columns for bonus
numeric_cols = ["diff_mean_bonus_percent","diff_median_bonus_percent"]
for column in numeric_cols:
  data[column] = data[column].fillna(data[column].median())
# drop irrelevant columns with missing values
data_cleaned = data.drop(columns=["company_link_to_gpg_info","responsible_person","post_code","address","company_number","sic_codes"])
# count latest missing values
print(data_cleaned.isnull().sum())
# print latest dataset size
print(f"data_cleaned shape:{data_cleaned.shape}")
```
