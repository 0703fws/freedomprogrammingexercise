```python
#Read/Summarise Data
# import python packages
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# read the data from the provided url
url ="https://raw.githubusercontent.com/0703fws/freedomprogrammingexercise/refs/heads/main/paygap.csv"
data = pd.read_csv(url)
# show the dataset
display(data)
def summarise_data(data):
    # check dataset details
    data.info()
    # summary statistics for numerical columns
    display(data.describe())
    # count missing values
    print(data.isnull().sum())
# check if data is loaded successfully
if data is not None:
    summarise_data(data)
else:
    print("data loading failed")
```

```python
#　Data Manipulation
# check if missing values position matches between male and female quartiles
overlap1 = data["male_lower_quartile"].isnull().equals(data["female_lower_quartile"].isnull())
overlap2 = data["male_lower_middle_quartile"].isnull().equals(data["female_lower_middle_quartile"].isnull())
overlap3 = data["male_upper_middle_quartile"].isnull().equals(data["female_upper_middle_quartile"].isnull())
overlap4 = data["male_top_quartile"].isnull().equals(data["female_top_quartile"].isnull())
# show matching results for each pair
overlap_result = {
    "male_lower_quartile vs female_lower_quartile":overlap1,
    "male_lower_middle_quartile vs female_lower_middle_quartile":overlap2,
    "male_upper_middle_quartile vs female_upper_middle_quartile":overlap3,
    "male_top_quartile vs female_top_quartile":overlap4
}
for group,match_result in overlap_result.items():
    print(f"{group}:{match_result}")
# define columns for each quartile
quartile_cols = ["male_lower_quartile","female_lower_quartile",
          "male_lower_middle_quartile","female_lower_middle_quartile",
          "male_upper_middle_quartile","female_upper_middle_quartile",
          "male_top_quartile","female_top_quartile"]
# fill missing values with median
for column in quartile_cols:
  data[column] = data[column].fillna(data[column].median())
# verify the sum of male-female pairs equals 100
pairs = [("male_lower_quartile","female_lower_quartile"),
    ("male_lower_middle_quartile","female_lower_middle_quartile"),
    ("male_upper_middle_quartile","female_upper_middle_quartile"),
    ("male_top_quartile","female_top_quartile")
]
for male,female in pairs:
  sum_pairs = data[male]+data[female]
  print(f"{male}+{female}:")
  print(f"sum:{sum_pairs.mean():.2f}")
# define columns for bonus
numeric_cols = ["diff_mean_bonus_percent","diff_median_bonus_percent"]
for column in numeric_cols:
  data[column] = data[column].fillna(data[column].median())
# drop irrelevant columns with missing values
data_cleaned = data.drop(columns=["company_link_to_gpg_info","responsible_person","post_code","address","company_number","sic_codes"])
# count latest missing values
print(data_cleaned.isnull().sum())
# print latest dataset size
print(f"data_cleaned shape:{data_cleaned.shape}")
```

```python
#Data visualisation: Pay gap by company size
# select columns for analysis
paygap_column = ["diff_mean_hourly_percent","diff_median_hourly_percent"]
# calculate mean pay gap grouped by company size
diff_size = data_cleaned.groupby("employer_size")[paygap_column].mean()
# create a bar chart
bar_chart = diff_size.plot(kind="bar",figsize=(10,6),edgecolor="black",width=0.8)
# add xlabel, ylabel and title
plt.xlabel("Company size",fontsize=12)
plt.ylabel("Pay gap percentage(%)",fontsize=12)
plt.title("Pay gap by company size",fontsize=18)
# rotate x-axis labels 45 degrees for easier reading and align them to the right
plt.xticks(rotation=45,ha="right")
# add legend and position outside the plot
plt.legend(["Mean Hourly Pay Gap","Median Hourly Pay Gap"],loc="upper left",bbox_to_anchor=(1,1))
# add value labels on bars
for container in bar_chart.containers:
  bar_chart.bar_label(container,fmt="%.1f%%",label_type="edge",fontsize=8)
# adjust layout
plt.tight_layout()
# show the bar chart
plt.show()
```

```python
#Data visualisation: Gender distribution by pay quartiles
# define quartile level labels
quartile_labels = ["Lower Quartile","Lower Middle Quartile","Upper Middle Quartile","Top Quartile"]
# create male quartile dictionary
male_quartile = {"male_lower_quartile":data_cleaned["male_lower_quartile"],
          "male_lower_middle_quartile":data_cleaned["male_lower_middle_quartile"],
          "male_upper_middle_quartile":data_cleaned["male_upper_middle_quartile"],
          "male_top_quartile":data_cleaned["male_top_quartile"]
}
# calculate mean values for male in each quartile
male_quartile_mean = [male_quartile["male_lower_quartile"].mean(),
     male_quartile["male_lower_middle_quartile"].mean(),
     male_quartile["male_upper_middle_quartile"].mean(),
     male_quartile["male_top_quartile"].mean()
]
# create female quartile dictionary
female_quartile = {"female_lower_quartile":data_cleaned["female_lower_quartile"],
          "female_lower_middle_quartile":data_cleaned["female_lower_middle_quartile"],
          "female_upper_middle_quartile":data_cleaned["female_upper_middle_quartile"],
          "female_top_quartile":data_cleaned["female_top_quartile"]
}
# calculate mean values for female in each quartile
female_quartile_mean = [female_quartile["female_lower_quartile"].mean(),
     female_quartile["female_lower_middle_quartile"].mean(),
     female_quartile["female_upper_middle_quartile"].mean(),
     female_quartile["female_top_quartile"].mean()
]
# set the size of the figure
plt.figure(figsize=(10,6))
# create a line chart for male quartile mean
plt.plot(quartile_labels,male_quartile_mean,label="Male",marker="o",linestyle="-",linewidth=2)
# add percentage labels for male data points
for i,value in enumerate(male_quartile_mean):
  plt.annotate(f"{value:.1f}%",(quartile_labels[i],value),textcoords="offset points",xytext=(0,6),ha="center") # offset above point
# create a line chart for female quartile mean
plt.plot(quartile_labels,female_quartile_mean,label="Female",marker="o",linestyle="-",linewidth=2)
# add percentage labels for female data points
for i,value in enumerate(female_quartile_mean):
  plt.annotate(f"{value:.1f}%",(quartile_labels[i],value),textcoords="offset points",xytext=(0,-12),ha="center") # offset below point
# add xlabel, ylabel and title
plt.xlabel("Quartiles",fontsize=12)
plt.ylabel("Percentage(%)",fontsize=12)
plt.title("Gender Distribution Across Quartiles",fontsize=18)
# add legend
plt.legend()
# show the line chart
plt.show()
```

```python
# Data visualisation: Proportion of employees receiving bonus by gender
# calculate mean bonus percentage
bonus_mean = data_cleaned[["male_bonus_percent","female_bonus_percent"]].mean()
# set the size of the figure
plt.figure(figsize=(10,6))
# create a pie chart with percentage labels for each segment
plt.pie(bonus_mean,labels=["Male","Female"],autopct="%1.1f%%")
# add title
plt.title("Proportion of Employees Receiving Bonus by Gender",fontsize=16)
# show the pie chart
plt.show()
```
